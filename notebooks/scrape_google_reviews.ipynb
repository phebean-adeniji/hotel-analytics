```python
def get_hotel_reviews(driver, url):  
    """"Fetch hotel reviews by providing their url."""

    driver.get(url) 
    time.sleep(5)
    SCROLL_PAUSE_TIME = 7

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
    number = 0

    while True:
        number = number+1  
        html = driver.find_element(By.TAG_NAME, 'html')
        html.send_keys(Keys.END)
        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)
        # Calculate new scroll height and compare with last scroll height
        print(f'last height: {last_height}')
        html = driver.find_element(By.TAG_NAME, 'html')
        html.send_keys(Keys.END)
        new_height = driver.execute_script("return arguments[0].scrollHeight", html)
        print(f'new height: {new_height}')   
        if (number != 500) and (last_height == new_height): 
            # time.sleep(3)
            html = driver.find_element(By.TAG_NAME, 'html')
            html.send_keys(Keys.UP)
            # time.sleep(1.5)
            html.send_keys(Keys.END)
        print('cont')
        last_height = new_height 
        if (number == 500) and (new_height == last_height): # if you want 60 samples, for ex let number = 7
            break
        print('cont')
        last_height = new_height  
    src = driver.page_source
    soup = BeautifulSoup(src, 'lxml')
    sections = soup.find_all('div', {'class': 'Svr5cf bKhjM'})

    return sections

driver = selenium_chrome_driver.access_chromdriver()
reviews_ = get_hotel_reviews(driver, lagos_oriental)
print(len(reviews_))
```

```python
# div class name for scrolling effect - fU5h8b YYZeDd
# div class name that contains each review box (so used find_elements) - Svr5cf bKhjM
# "a" tag class name that contains reviewer names - DHIhE QB2Jof
# div class name that contains the review texts (find_elements and take the last one because it contains the full review) - K7oBsc
# div class name that contains the ratings - GDWaad

img_url_list = []
rating_list = []
review_list = []
image_list = []

for item in reviews_:
    review = item.find_all('div', {'class': 'K7oBsc'})
    ratings = item.find('div', {'class': 'GDWaad'})
    
    rating_list.append(ratings.text)
    review_list.append(review[-1].text)

    img = item.find_all('img', {'class': 'x7VXS vmf5Wb inc9Wb VP1UCc'}) # img could be empty

    try:
        image = [i["src"] for i in img] # extract image urls
        print(image, "\n")
    except:
        image = []
    image_list.append(image)
    
```

```python
review_df = pd.DataFrame(
    {'rating': rating_list,
     'review': review_list,
     'image_url': image_list})

review_df.head()
```
